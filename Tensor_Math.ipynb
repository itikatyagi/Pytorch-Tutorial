```python
import torch
import numpy as np
```

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
## Add
tensor_a + tensor_b
```

```python
## Addition longhand
torch.add(tensor_a, tensor_b)
```

```python
## subtraction
tensor_b - tensor_a
```

```python
## sub function
torch.sub(tensor_b, tensor_a)
```

```python
## multiplication
tensor_a * tensor_b
```

```python
## mul function
torch.mul(tensor_a, tensor_b)
```

```python
## Divide
tensor_b / tensor_a
```

```python
## Divide function
torch.div(tensor_b,tensor_a)
```

```python
## Remainder Modulus
tensor_b % tensor_a
```

```python
## Remainder longhand
torch.remainder(tensor_b, tensor_a)
```

```python
## Exponents/Power
torch.pow(tensor_a, tensor_b)
```

```python
## Another way to write add
tensor_a.add(tensor_b)
```
